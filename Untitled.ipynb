```python
import pandas as pd
%matplotlib inline
import pylab as plt
import seaborn as sns
import numpy as np
import h2o
from h2o.automl import H2OAutoML
```

```python
sample=pd.read_csv('sample.csv')
sample.head()
```

```python
train=pd.read_csv('train.csv')
train.head()
```

```python
train.info()
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(train.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)



sns.heatmap(train.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});

```

```python
test=pd.read_csv('test.csv')
```

```python
h2o.init()
```

```python
# parsear datos para h20

h2train=h2o.H2OFrame(train)
h2test=h2o.H2OFrame(test)

```

```python
h2train.columns
```

```python
X=[c for c in h2train.columns if c!='price']

y='price'
```

```python
automl=H2OAutoML(max_models=50,
                 seed=42,   # random_state
                 max_runtime_secs=300,
                 sort_metric='RMSE')
```

```python
automl.train(x=X,
             y=y,
             training_frame=h2train)
```

```python
print('[INFO] Leader board:')

leader_board=automl.leaderboard

leader_board.head()
```

```python
y_pred=automl.leader.predict(h2test)
```

```python
pred= h2o.as_list(y_pred)
len(pred)
```

```python
sample['price']=pred
```
