```python
columns=['property_type','neighbourhood_cleansed','host_location','availability_30','accommodates','host_verifications','bedrooms','availability_365','host_acceptance_rate', 'price']
```

```python
datos=train[columns]
datos.info()  
```

```python
datos.dropna(inplace=True)
datos.reset_index(inplace=True)
```

```python
datos.drop(columns='index', inplace=True)
datos
```

```python
neighbourhood_cleansed=datos.neighbourhood_cleansed.unique()
neighbourhood_cleansed
```

```python
precio=[datos[datos['neighbourhood_cleansed']==i]['price'].mean() for i in neighbourhood_cleansed]
barrios=pd.DataFrame(neighbourhood_cleansed)
barrios['precio']=precio
barrios.rename(columns={0: "barrio"}, inplace=True)
```

```python
barrios['precio']=barrios['precio']/barrios['precio'].min()
#barrios['std']=[data[data['neighbourhood_cleansed']==i]['price'].std() for i in neighbourhood_cleansed]
barrios.head()
```

```python
n=barrios.set_index('barrio').T.to_dict('list')
n
```

```python
datos['neighbourhood_cleansed']=datos['neighbourhood_cleansed'].map(n)

datos.neighbourhood_cleansed=[float(datos.neighbourhood_cleansed[i][0]) for i in range(len(datos))]
datos
```

```python
datos.host_acceptance_rate=datos.host_acceptance_rate.replace('%',' ')
datos.info()
```
